# PyCity Schools Analysis

- Overall Analysis: This data set was based on two CSV files of approximately 39000 students who are attending 14 different high schools.  This analysis sought to understand and visualize the relative sucess rate for these schools, first as a whole, and then displayed on a "per school" basis.  The KPIs that were measured involved the scores (base, average, and % of population) for math and reading.  As well as the overall success rate based on students who passed both subjects.  This data was further compared to the total budget of these schools and the per student spending.  With additional visualization used to denote whether the school was a charter school, or a district school.  Once these visualizations had occurred, I analyzed the data and drew two conclusions that I felt were relevant.

- Observation 1:  There seems to be a correlation between a larger population and a decline in the overall % passing rate for the schools in this Data Set.  In 7 of the 8 schools classified as large (hereafter defined as a school whose population is > 2000), overall passing % didn't get above 55%.  Wilson Highschool is the outlier, it has a population of 2283, and has an overall passing % of 90%.  The next lowest population of the "large" schools is Ford Highschool.  Ford has a population of 2739 and has an overall passing %  of 54%.  This is in keeping with other schools in it's classification.  It would stand to reason that once the population surpasses 2500, a decline in overall passing % may be seen.  There is also a direct correlation between schools that are considered to be district schools vs schools that are considered charter school.  All district schools in this dataset are classified as large schools.  The only charter school that is considered large was Wilson Highschool which was shown to have test scores in line with other Charter schools as was already noted.

  
- Observation 2: It was noted that budget seemed to have little consequence on the relative success rate of the students attending these schools.  The schools with the highest budget were the district schools, and they also had the highest per student spending.  Yet as was already noted, they were all in the bottom 5 with regards to test scores.  Meanwhile the top 5 schools were all having around 1 Million dollars, with the exception of Pena High School which is number 5 on the top 5, but only had a budget of 500 thousand.  It should be noted however, that Pena High School has the second lowest population for the entire data set.  As a result their per student spending is moderate.

```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the average (mean) math score
average_math_score = pd.DataFrame(student_data)[("math_score")].mean().round(3)
average_math_score



print("The mean math score for all students was", average_math_score)
```

```python
# Calculate the average (mean) reading score
average_reading_score = pd.DataFrame(student_data)[("reading_score")].mean().round(3)
average_reading_score

print("The mean reading score for all students was", average_reading_score)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(round(passing_math_percentage,3))
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100

print(round(passing_reading_percentage,3))
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(round(overall_passing_rate,3))
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_summary = [{"Unique Schools" : school_count, "Total Students" : student_count, "Total Budget" : total_budget,
                     "Average Math Score" : average_math_score, "Average Reading Score" : average_reading_score, "% Passing Math" : round(passing_math_percentage,3),
                     "% Passing Reading" : round(passing_reading_percentage,3), "% Overall Passing" : round(overall_passing_rate,3)}]

dis_sum = pd.DataFrame(district_summary)

# Display the DataFrame

dis_sum
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby('school_name')['math_score'].mean().reset_index()
tenth_grader_math_scores = tenth_graders.groupby('school_name')['math_score'].mean().reset_index()
eleventh_grader_math_scores = eleventh_graders.groupby('school_name')['math_score'].mean().reset_index()
twelfth_grader_math_scores = twelfth_graders.groupby('school_name')['math_score'].mean().reset_index()

# # Combine each of the scores above into single DataFrame called `math_scores_by_grade`

# Note: multiple merges tended to cause issues.  Doing 9th and 10th, then 11th and 12th, then merging all together went smooth
ninth_tenth_merge_math = pd.merge(ninth_grade_math_scores, tenth_grader_math_scores, on=['school_name'], how = 'inner')
eleventh_twelfth_merge_math = pd.merge(eleventh_grader_math_scores, twelfth_grader_math_scores, on = ['school_name'], how = 'inner')


math_scores_by_grade= pd.merge(ninth_tenth_merge_math, eleventh_twelfth_merge_math, on= ['school_name'], how = 'inner')

math_scores_by_grade.rename(columns = {'math_score_x_x': 'Ninth Grade Math Scores', 'math_score_y_x': "Tenth Grade Math Scores", "math_score_x_y": "Eleventh Grade Math Scores", 'math_score_y_y' : "Twelfth Grade Math Scores"}, inplace=True)

# # Minor data wrangling
math_scores_by_grade.index.name = None

# # Display the DataFrame
math_scores_by_grade

```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby('school_name')['reading_score'].mean().reset_index()
tenth_grader_reading_scores = tenth_graders.groupby('school_name')['reading_score'].mean().reset_index()
eleventh_grader_reading_scores = eleventh_graders.groupby('school_name')['reading_score'].mean().reset_index()
twelfth_grader_reading_scores = twelfth_graders.groupby('school_name')['reading_score'].mean().reset_index()

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`

ninth_tenth_merge_reading = pd.merge(ninth_grade_reading_scores, tenth_grader_reading_scores, on=['school_name'], how = 'inner')
eleventh_twelfth_merge_reading = pd.merge(eleventh_grader_reading_scores, twelfth_grader_reading_scores, on = ['school_name'], how = 'inner')


reading_scores_by_grade = pd.merge(ninth_tenth_merge_reading, eleventh_twelfth_merge_reading, on= ['school_name'], how = 'inner')

reading_scores_by_grade.rename(columns = {'reading_score_x_x': 'Ninth Grade Reading Scores', 'reading_score_y_x': "Tenth Grade Reading Scores", "reading_score_x_y": "Eleventh Grade Reading Scores", 'reading_score_y_y' : "Twelfth Grade Reading Scores"}, inplace=True)

# Minor data wrangling
# reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

```python
type(school_spending_df["Spending Ranges (Per Student)"])
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["math_score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["reading_score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary_calc = {"Average Math Spending": spending_math_scores, "Average Reading Spending": spending_reading_scores,
                   "Students Passing Math": spending_passing_math, "Students Passing Reading": spending_passing_reading,
                   "Overall Passing Spending" : overall_passing_spending}

spending_summary = pd.DataFrame(spending_summary_calc)

# Display results
spending_summary
```
